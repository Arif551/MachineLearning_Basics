```python
import matplotlib.pyplot as plt
%matplotlib inline
def gradient_decent(x,y):
    m = b = 0
    lr = 0.09
    it = 1000
    n = len(y)
    for i in range(it):
        y_predict = m*x + b
        mse = (1/n) * sum([s**2 for s in (y-y_predict)])
        md = (-2/n) * sum(x * (y-y_predict))
        bd = (-2/n) * sum(y-y_predict)
        b = b - lr * bd
        m = m - lr * md
        plt.scatter(x,y_predict)
        print("m {} , b {} , cost {} iteration {}".format(m,b,mse,i))
```

```python
import numpy as np
x = np.array([1,2,3,4,5,7])
y = np.array([12,14,16,18,20,22])
```

```python
gradient_decent(x,y)
```
