```python
import pandas as pd
df = pd.read_csv('C:/Users/ARIF RAJA MONDAL/Desktop/Machine Learning/Decision Tree/salarydata.csv')
df
```

```python
input = df.drop(['salary>=100k'],axis=1)
input
type(input)
```

```python
target = df['salary>=100k']
target
type(target)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
input['company_n'] = le_company.fit_transform(input['company'])
input['job_n'] = le_job.fit_transform(input['job'])
input['degree_n'] = le_degree.fit_transform(input['degree'])
input
```

```python
input_n = input.drop(['company','job','degree'], axis =1)
input_n
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model
```

```python
model.fit(input_n,target)
```

```python
model.score(input_n,target)
```

```python
model.predict([[0,1,1]])
```
